```python
import csv
import numpy
import matplotlib.pyplot as plt

CO2_auto_correlation = []
temp_auto_correlation = []
cross_correlation = []
```

```python
avg_CO2 = []
avg_temp = []

with open("samples.csv", 'rU') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    first = True

    for row in reader:
        if first == None:
            avg_co2_temp = 0
            if row[2] == 70:
                avg_co2_temp = 0
            else:
                avg_co2_temp = (float(row[3]) + float(row[4])) / 2
            avg_CO2.append(avg_co2_temp)
            avg_temp.append((float(row[5]) + float(row[6])) / 2)
        first = None
```

```python
def auto_correlate(sample):
    result=numpy.correlate(sample,sample,"same")
    return result
```

```python
def cross_correlate(sample1, sample2):
    result=numpy.correlate(sample1,sample2,"same")
    return result
```

```python
CO2_auto_correlation=auto_correlate(avg_CO2)
temp_auto_correlation=auto_correlate(avg_temp)
cross_correlation=cross_correlate(avg_CO2,avg_temp)
```

```python
def CO2_auto_correlate():
    fig = plt.figure()
    fig.suptitle('Auto-correlation of CO2 samples', fontsize=20)
    plt.plot(avg_CO2)
    plt.plot(auto_correlate(avg_CO2))
    plt.show()
```

```python
def temparature_auto_correlate():

    fig = plt.figure()
    fig.suptitle('Auto-correlation of temperature Samples', fontsize=20)
    plt.ylabel('Magnitude', fontsize=16)
    plt.plot(auto_correlate(avg_temp))
    plt.show()
```

```python
def cross_correlate_waves():
    fig = plt.figure()
    fig.suptitle('Cross-correlation between CO2 and temperature samples', fontsize=20)
    plt.ylabel('Magnitude', fontsize=16)
    plt.plot(cross_correlate(avg_CO2,avg_temp))
    plt.show()
```

```python
def plotCO2():
    fig = plt.figure()
    fig.suptitle('Avg CO2 samples', fontsize=20)
    plt.ylabel('Magnitude', fontsize=16)
    plt.plot(avg_CO2)
    plt.show()
```

```python
def plotTemperature():
    fig = plt.figure()
    fig.suptitle('Avg temperature samples', fontsize=20)
    plt.plot(avg_temp)
    plt.ylabel('Magnitude', fontsize=16)
    plt.show()
```

```python
plotCO2()
CO2_auto_correlate()

plotTemperature()
temparature_auto_correlate()


cross_correlate_waves()
```
